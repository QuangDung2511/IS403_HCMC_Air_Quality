```python
# Import lib

import pandas as pd
import os
```

```python
# Load the datasets

air_df = pd.read_csv('data/hcm_air_data.csv')

weather_df = pd.read_csv('data/hcmc_weather_data.csv')
```

```python
# Convert to datetime objects

air_df['datetime_local'] = pd.to_datetime(air_df['datetime_local'])

weather_df['time'] = pd.to_datetime(weather_df['time'])
```

```python
# Remove the timezone

air_df['datetime_local_naive'] = air_df['datetime_local'].dt.tz_localize(None)
```

```python
# Merge the datasets

df = pd.merge(
    air_df, 
    weather_df, 
    left_on='datetime_local_naive', 
    right_on='time', 
    how='left'
)
```

```python
# Clean up

df = df.drop(columns=['datetime_local_naive', 'time'])

df = df.sort_values(by='datetime_local').reset_index(drop=True)
```

```python
df = df.set_index('datetime_local')
```
