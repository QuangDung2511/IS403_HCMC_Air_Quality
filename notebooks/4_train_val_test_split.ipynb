# Train / Val / Test Split

```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("../data/processed/hcmc_features_continuous.csv")
OUT_DIR = Path("../data/processed")

df = pd.read_csv(DATA_PATH)
df["datetime_local"] = pd.to_datetime(df["datetime_local"])
df = df.sort_values("datetime_local").reset_index(drop=True)

print(f"Total Rows: {len(df)}")
```

```python
n = len(df)
i1 = int(n * 0.6)
i2 = int(n * 0.8)

train_df = df.iloc[:i1].copy()
val_df = df.iloc[i1:i2].copy()
test_df = df.iloc[i2:].copy()

for name, part in ("train", train_df), ("val", val_df), ("test", test_df):
    print(f"{name}: n={len(part)}, datetime_local [{part['datetime_local'].min()} .. {part['datetime_local'].max()}]")

assert len(train_df) + len(val_df) + len(test_df) == n
```

```python
train_df.to_csv(OUT_DIR / "train_split.csv", index=False)
val_df.to_csv(OUT_DIR / "val_split.csv", index=False)
test_df.to_csv(OUT_DIR / "test_split.csv", index=False)

print("Splits saved successfully!")
```
