# Train, Validation, Test Split

This notebook splits the continuous highly engineered features into train, val, and test splits (60/20/20 chronological) to prevent data leakage.

```python
import pandas as pd
import os

df = pd.read_csv('../data/processed/hcmc_features_continuous.csv', index_col=0)
df.index = pd.to_datetime(df.index)
print(f"Total shape: {df.shape}")
```

```python
train_size = int(len(df) * 0.6)
val_size = int(len(df) * 0.2)

train_df = df.iloc[:train_size]
val_df = df.iloc[train_size:train_size+val_size]
test_df = df.iloc[train_size+val_size:]

print(f"Train shape: {train_df.shape}")
print(f"Val shape: {val_df.shape}")
print(f"Test shape: {test_df.shape}")
```

```python
train_df.to_csv('../data/processed/train_split.csv')
val_df.to_csv('../data/processed/val_split.csv')
test_df.to_csv('../data/processed/test_split.csv')
print('Data successfully split and saved.')
```
